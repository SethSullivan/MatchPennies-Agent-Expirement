```python
time = np.arange(0,1800,1)
mu1 = 1100
mu2 = 1000
dist1 = stats.norm.pdf(time,mu1,150)
dist2 = stats.norm.pdf(time,mu2,150)
fig,ax = plt.subplots(dpi=300,figsize=(16,6))
ax.plot(dist1,c=wheel.purple,linewidth=5)
ax.plot(dist2,c=wheel.sunflower,linewidth=5)
ax.axvline(1500,color='white',lw = 6)
ax.fill_between(time[1500:],0,dist2[1500:],color='grey')

ax.text(0.75,1.1,'Time\nConstraint',verticalalignment = 'center',ha='center',rotation = 0,transform=ax.transAxes, fontsize=28, fontweight='bold')
ax.text(mu1+35,max(dist1)+0.00015,r'$\mathbf{\mu}$='+f'{mu1}',va='center',ha='center',c=wheel.purple,fontweight='bold', fontsize=25
        )
ax.text(mu2-35,max(dist2)+0.00015,r'$\mathbf{\mu}$='+f'{mu2}',va='center',ha='center',c=wheel.sunflower,fontweight='bold',fontsize=25)
ax.set_yticks([])
ax.spines['left'].set_visible(False)
ax.set_xlim(600,1800)
xlocs = np.arange(600,2100,300)
ax.set_xticks(xlocs)
ax.set_xticklabels(xlocs, fontsize=30)
ax.set_ylim(0,max(dist1)+0.0001)
ax.set_xlabel('Target Reach Time (ms)', fontsize=35)
dv.custom_legend(ax,['Predator','Prey'],[wheel.sunflower,wheel.purple],loc='upper left',fontsize=25)
```

```python
#
#* Plot Parameters
timesteps = np.arange(0,2000,1)
mu1,mu2,mu3 = 1000,1100,1200
sd1,sd2 = 50,150

dist1 = stats.norm.pdf(timesteps,mu1,sd1) 
dist2 = stats.norm.pdf(timesteps,mu1,sd2) 
dist3 = stats.norm.pdf(timesteps,mu2,sd1) 
dist4 = stats.norm.pdf(timesteps,mu2,sd2) 
dist5 = stats.norm.pdf(timesteps,mu3,sd1) 
dist6 = stats.norm.pdf(timesteps,mu3,sd2)

dists = [dist1, dist2, dist3, dist4, dist5, dist6]
colors = constants.xticklabel_colors_sd
legend_labels = [
        r'$\mathbf{\mu}$' + f' = {mu1}\n' + r'$\mathbf{\sigma}$' + f' = {sd1}', r'$\mathbf{\mu}$' + f' = {mu1}\n' + r'$\mathbf{\sigma}$' + f' = {sd2}',
        r'$\mathbf{\mu}$' + f' = {mu2}\n' + r'$\mathbf{\sigma}$' + f' = {sd1}', r'$\mathbf{\mu}$' + f' = {mu2}\n' + r'$\mathbf{\sigma}$' + f' = {sd2}',
        r'$\mathbf{\mu}$' + f' = {mu3}\n' + r'$\mathbf{\sigma}$' + f' = {sd1}', r'$\mathbf{\mu}$' + f' = {mu3}\n' + r'$\mathbf{\sigma}$' + f' = {sd2}',
]
######### # All Plots, One plot,   Just 2 sds, All means, one sd 
slices = [slice(None), slice(2,3), slice(2,4), slice(0,-1,2)]
save_names = [
    'approach_panel_all_distributions', 'approach_panel_one_distribution',
    'approach_panel_two_sds','approach_panel_all_means'
]
legend_locs = [
    'lower center','center',
    'lower center', 'upper center'
]
bbox_y_anchor = [
    0.92, 1.302,
    0.92, 1.465
]
# plot_distributions = [dists, dists[2:2], dists[2:4], dists[::2]]
# colors_list = [colors, colors[2:2], colors[2:4], colors[::2]]
# legend_labels = [colors, colors[2:2], colors[2:4], colors[::2]]

#* All Plots
for i in range(len(slices)):
    fig = dv.Figure(figsize=(4.67,2.83),dpi= 300)
    ax = fig.add_panel((0.15,.67,4.3,1.69))
    num_plots = len(dists[slices[i]])
    for j in range(num_plots):
        ax.plot(dists[slices[i]][j],color=colors[slices[i]][j],lw=5)

    if num_plots <3:
        ncols = 1
    else:
        ncols = 3
    
    dv.legend(ax,labels=legend_labels[slices[i]], 
                colors=colors[slices[i]], fontsize=10, ncols=ncols, 
                loc=legend_locs[i],bbox_to_anchor=(0.5,bbox_y_anchor[i]),
                handlelength=0.93, columnspacing=2.5,lw=7 ),
    ax.set_xlim(600,1600)
    ax.get_yaxis().set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.tick_params(labelsize=10)
    ax.set_xlabel('Agent Movement Onset Time (ms)', fontsize=12)
    ax.set_facecolor('none')
    plt.show()
    fig.savefig(SAVE_PATH / save_names[i])
```
