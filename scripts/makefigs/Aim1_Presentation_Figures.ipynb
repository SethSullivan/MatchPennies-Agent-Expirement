```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import data_visualization as dv
wheel = dv.ColorWheel()
plt.style.use('cashaback_dark')
```

```python
time = np.arange(0,1800,1)
dist = stats.norm.pdf(time,1000,150)
```

```python
stats.norm.cdf(750,1000,150)
```

```python
stop_time = 750
fig,ax = plt.subplots(dpi=300,figsize=(10,4))
ax.plot(dist,color=wheel.grey,linewidth=8.0)
ax.set_yticks([])
ax.spines['left'].set_visible(False)
ax.fill_between(time[:stop_time],0,dist[:stop_time],color=wheel.light_orange,alpha=0.8)
ax.fill_between(time[stop_time:],0,dist[stop_time:],color = wheel.pink,alpha=0.8)
ax.set_xlabel('Time (ms)')
```

```python
time = np.arange(0,1800,1)
mu1 = 1100
mu2 = 1300
dist1 = stats.norm.pdf(time,mu1,150)
dist2 = stats.norm.pdf(time,mu2,150)
fig,ax = plt.subplots(dpi=300,figsize=(10,4))
ax.plot(dist1,c=wheel.purple,linewidth=3)
ax.plot(dist2,c=wheel.sunflower,linewidth=3)
ax.axvline(1500,color='white',lw = 3)
ax.fill_between(time[1500:],0,dist2[1500:],color='grey')

ax.text(0.76,0.7,'Time Constraint',verticalalignment = 'center',ha='center',rotation = -90,transform=ax.transAxes)
ax.text(mu1-20,max(dist1)+0.0001,r'$\mathbf{\mu}$='+f'{mu1}',va='center',ha='left',c=wheel.purple,fontweight='bold')
ax.text(mu2-60,max(dist2)+0.0001,r'$\mathbf{\mu}$='+f'{mu2}',va='center',ha='left',c=wheel.sunflower,fontweight='bold')
ax.set_yticks([])
ax.spines['left'].set_visible(False)
ax.set_xlim(600,1800)
ax.set_xticks(np.arange(600,2100,300))
ax.set_ylim(0,max(dist1)+0.0001)
ax.set_xlabel('Target Reach Time (ms)')
dv.custom_legend(ax,['Predator','Prey'],[wheel.sunflower,wheel.purple],loc='upper left',fontsize=10)
```
