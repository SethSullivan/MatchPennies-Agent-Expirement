```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

```python
myclip_a = 1100
myclip_b = 1200 # BUT THIS CLIP TIME IS PROBABILISTIC as WELL
my_mean = 1200
my_std = 50

# Just inputtting myclip_a and myclip_b determines how many stds away to clip
# The below equations convert to the actual CLIP SPOTS IN THE SAMPLE SPACE
a, b = (myclip_a - my_mean) / my_std, (myclip_b - my_mean) / my_std
x = np.arange(1000,2000,1)

y = stats.truncnorm.pdf(x,a,b,loc=my_mean,scale=my_std)
fig,ax = plt.subplots()
ax.plot(x,y)
```
