```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
myclip_a = 0
myclip_b = 1100 # BUT THIS CLIP TIME IS PROBABILISTIC as WELL
my_mean = 1500
my_std = 50

# Just inputtting myclip_a and myclip_b determines how many stds away to clip
# The below equations convert to the actual CLIP SPOTS IN THE SAMPLE SPACE
a, b = (myclip_a - my_mean) / my_std, (myclip_b - my_mean) / my_std

x = stats.truncnorm.cdf(np.arange(0,2000,1),a,b,loc=my_mean,scale=my_std)
fig,ax = plt.subplots()
ax.plot(x)
ax.set_xlim(1000,1150)
```

```python
y = stats.norm.pdf(np.arange(0,2000,1),loc=1000,scale=50)
fig,ax = plt.subplots()
ax.plot(y)
```
