## Simulation Functions

```python
#%% Simulation Function Only player Sensory Feedback
@njit 
def Simulate_Wins_Only_Player_Sensory_Feedback(TRIAL_TIME,NUM_TRIALS,REACTION_TIME,MOVEMENT_TIME, decision_means_player, decision_means_agent,
                  player_decision_times, agent_decision_times):
    # Initialize what I want to return
    player_wins = np.zeros((len(decision_means_player), len(decision_means_agent)))
    agent_wins = np.zeros((len(decision_means_player), len(decision_means_agent)))
    player_indecisions = np.zeros((len(decision_means_player), len(decision_means_agent)))
    player_reaction_indecisions = np.zeros((len(decision_means_player), len(decision_means_agent)))
    player_gamble_indecisions = np.zeros((len(decision_means_player), len(decision_means_agent)))
    agent_indecisions = np.zeros((len(decision_means_player), len(decision_means_agent)))
    neither_decision = np.zeros((len(decision_means_player), len(decision_means_agent)))
    player_decision_array = np.empty((len(decision_means_player), len(decision_means_agent), NUM_TRIALS)) #0 is indecision, 1 is win, -1 is loss
    agent_decision_array = np.empty((len(decision_means_player), len(decision_means_agent), NUM_TRIALS))
    player_both_decided_array = np.empty((len(decision_means_player), len(decision_means_agent), NUM_TRIALS)) #0 is indecision, 1 is win, -1 is loss
    agent_both_decided_array = np.empty((len(decision_means_player), len(decision_means_agent), NUM_TRIALS))
    player_reach_times = np.empty((len(decision_means_player), len(decision_means_agent), NUM_TRIALS))
    player_leave_target_times = np.empty((len(decision_means_player), len(decision_means_agent), NUM_TRIALS))
    agent_reach_times = np.empty((len(decision_means_player), len(decision_means_agent), NUM_TRIALS))
    
    player_reaction_wins = np.zeros((len(decision_means_player), len(decision_means_agent)))

    player_gamble_wins = np.zeros((len(decision_means_player), len(decision_means_agent)))
    player_gamble_incorrects = np.zeros((len(decision_means_player), len(decision_means_agent)))
    player_incorrects = np.zeros((len(decision_means_player), len(decision_means_agent)))
    c = -1 #decision time iterator
        # Simulation -------------------------------------------------------------
    for dt_player in (decision_means_player):
        c=c+1
        z=-1
        for dt_agent in decision_means_agent:
            z=z+1
            for i in range(NUM_TRIALS): # can probably speed this up by doing the norm.rvs for the number of trials instead of running through a for loop, but not sure how to do the if statements then                 
                # -------------Using sensory evidence ------------------------------------------
                # Set reach times to be without reaction, this is general, and get's changed only for the first if statement, so I put it at the top
                player_reach_times[c,z,i] = player_decision_times[c,z,i]+MOVEMENT_TIME
                agent_reach_times[c,z,i] = agent_decision_times[c,z,i]+AGENT_MOVEMENT_TIME
                if ((player_decision_times[c,z,i] - agent_decision_times[c,z,i] >= 0)): # EVEN if player is over the trial time, if they select a decision time that's after agent movement onset, they will react
                    player_leave_target_times[c,z,i] = agent_decision_times[c,z,i] + REACTION_TIME # change decision time to reaction time after Agent
                    player_reach_times[c,z,i] = player_leave_target_times[c,z,i] + MOVEMENT_TIME # Set reach time to include reaction 
                    if agent_decision_times[c,z,i] + REACTION_TIME + MOVEMENT_TIME < TRIAL_TIME: # IF We can respond to the agent in time we win
                        player_wins[c,z] += 1
                        player_reaction_wins[c,z] +=1
                        
                        player_decision_array[c,z,i] = 1
                        agent_decision_array[c,z,i] = -1
                        player_both_decided_array[c,z,i] = 1
                        agent_both_decided_array[c,z,i] = -1
                    else:
                        player_indecisions[c,z]+=1
                        player_reaction_indecisions+=1
                        
                # ------ No sensory evidence could be used------------------------
                
                # Both made it, random winner
                elif (player_decision_times[c,z,i]+MOVEMENT_TIME < TRIAL_TIME and
                      agent_decision_times[c,z,i]+AGENT_MOVEMENT_TIME <TRIAL_TIME): 
                    rand = random.random()
                    
                    if rand>=0.5:
                        player_wins[c,z] += 1
                        player_gamble_wins[c,z]+=1
                        player_decision_array[c,z,i] = 1
                        agent_decision_array[c,z,i] = -1
                        player_both_decided_array[c,z,i] = 1
                        agent_both_decided_array[c,z,i] = -1
                    else:
                        agent_wins[c,z] += 1
                        player_incorrects[c,z] += 1    
                        player_gamble_incorrects[c,z] += 1
                        player_decision_array[c,z,i] = -1
                        agent_decision_array[c,z,i] = 1     
                        player_both_decided_array[c,z,i] = -1
                        agent_both_decided_array[c,z,i] = 1                                                                                      
                # player didn't make it
                elif (player_decision_times[c,z,i]+MOVEMENT_TIME>= TRIAL_TIME and
                      agent_decision_times[c,z,i]+AGENT_MOVEMENT_TIME <TRIAL_TIME):
                    agent_wins[c,z] += 1
                    player_indecisions[c,z] += 1
                    player_gamble_indecisions[c,z]+=1
                    player_decision_array[c,z,i] = 0
                    agent_decision_array[c,z,i] = 1
                    player_both_decided_array[c,z,i] = 0
                    agent_both_decided_array[c,z,i] = 1
                # agent didn't make it
                elif (player_decision_times[c,z,i]+MOVEMENT_TIME<TRIAL_TIME and 
                      agent_decision_times[c,z,i]+AGENT_MOVEMENT_TIME>=TRIAL_TIME):
                    player_wins[c,z] += 1
                    player_gamble_wins[c,z] += 1
                    agent_indecisions[c,z] += 1
                    player_decision_array[c,z,i] = 1
                    agent_decision_array[c,z,i] = 0
                    player_both_decided_array[c,z,i] = 1
                    agent_both_decided_array[c,z,i] = 0
                # Both didn't make it 
                elif (player_decision_times[c,z,i]+MOVEMENT_TIME>=TRIAL_TIME and 
                      agent_decision_times[c,z,i]+AGENT_MOVEMENT_TIME>=TRIAL_TIME):
                    player_indecisions[c,z] += 1
                    agent_indecisions[c,z] += 1
                   
                    player_gamble_indecisions[c,z] += 1
                    neither_decision[c,z] += 1
                    player_decision_array[c,z,i] = 0
                    agent_decision_array[c,z,i] = 0
                    player_both_decided_array[c,z,i] = 0
                    agent_both_decided_array[c,z,i] = 0
                else:
                    print("ERROR NO IF STATEMENT TRIGGERED")     
    return player_wins, agent_wins, player_indecisions, agent_indecisions, player_incorrects, player_reach_times, agent_reach_times, player_decision_times,agent_decision_times, \
        player_decision_array, agent_decision_array, player_both_decided_array, agent_both_decided_array         
```

```python
#%% Calculate both decided wins
@njit
def calc_both_decide_wins(NUM_TRIALS, target_reach_means_p1,agent_decision_means,
                          p1_decision_array,
                          p2_decision_array, 
                          p1_both_decided_array, 
                          p2_both_decided_array):
    p1_both_decided_wins = np.zeros((len(target_reach_means_p1), len(agent_decision_means)))
    p2_both_decided_wins = np.zeros((len(target_reach_means_p1), len(agent_decision_means)))
    p1_perc_both_decided_wins = np.zeros((len(target_reach_means_p1), len(agent_decision_means)))
    p2_perc_both_decided_wins = np.zeros((len(target_reach_means_p1), len(agent_decision_means)))
       
    # p1_both_decided_array = p1_decision_array #decision array
    # p2_both_decided_array = p2_decision_array     
    p1_indecision_index = np.argwhere(p1_decision_array == 0)
    p2_indecision_index = np.argwhere(p2_decision_array == 0)
      # Create arrays for when both decide 
      #Replace all 0's (indecisions) with nans
    for i,j,k in p1_indecision_index:
        p1_both_decided_array[i,j,k] = np.nan
        p2_both_decided_array[i,j,k] = np.nan
    for i,j,k in p2_indecision_index:
       p1_both_decided_array[i,j,k] = np.nan
       p2_both_decided_array[i,j,k] = np.nan
    
    # Count wins for each when both decide
    for i in range(len(target_reach_means_p1)):
        for j in range(len(agent_decision_means)):
            for k in range(NUM_TRIALS):
                if p1_both_decided_array[i,j,k]==1 and p2_both_decided_array[i,j,k] == -1:
                    p1_both_decided_wins[i,j] += 1
                    # if np.isnan(p1_both_decided_array[i]):
                    #     print('error')
                    #print(f"Right wins on correct choice: {decision_array_R[i,j,k]}, {decision_array_L[i,j,k]}")
                # Prey won for correct choice
                elif p2_both_decided_array[i,j,k]==1 and p1_both_decided_array[i,j,k] == -1:
                    p2_both_decided_wins[i,j] += 1
                    
            x = np.count_nonzero(~np.isnan(p1_both_decided_array[i,j,:]))
            y = np.count_nonzero(~np.isnan(p2_both_decided_array[i,j,:]))
            if x != 0 and y != 0:
                total = np.count_nonzero(np.logical_and(p1_decision_array[i,j,:]!=0,p2_decision_array[i,j,:]!=0))
                p1_perc_both_decided_wins[i,j] = (p1_both_decided_wins[i,j] / total)*100
                p2_perc_both_decided_wins[i,j] = (p2_both_decided_wins[i,j] / total)*100
            else:
                p1_perc_both_decided_wins[i,j] = np.nan
                p2_perc_both_decided_wins[i,j] = np.nan
        
    return p1_perc_both_decided_wins, p2_perc_both_decided_wins

```
